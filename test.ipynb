```python
import csv

# Create an empty dictionary to store the word counts
word_counts = {}

# Open the CSV file
with open('output.csv', 'r') as file:

    # Create a CSV reader object
    reader = csv.reader(file)

    # Iterate over each row in the CSV file
    for row in reader:

        # Split the row's data by space unit
         for word in row[3].lower().strip().split():
            ## 단어가 dictionary에 등록 되어 있다면 카운트를 올려주고
            if word in word_counts:
                word_counts[word] += 1
            ## 단어가 dictionary에 없다면 등록하고 카운트를 1로 등록합니다.
            else:
                word_counts[word] = 1
   

# Print the word counts dictionary
word_list = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
print(word_list)
```
